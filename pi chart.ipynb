```python
#SET STYLE FOR REST OF NOTEBOOK
matplotlib.style.use('ggplot') #set stylesheet (find options here: https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)
matplotlib.rcParams['font.size'] = 27
matplotlib.rcParams['figure.figsize'] = (30, 30) #standardise figure size
matplotlib.rcParams['font.family'] =  'Verdana'
#matplotlib.rcParams['font.weight'] =  'bold' #<-consider adding this line (uncommented) if using a thin font#SET STYLE FOR REST OF NOTEBOOK
```

```python
def percentage(row, column2):
    totalsum = 0 #function for choosing what you are finding the perecntage of 

    qgwjsbgkjbkasbg = ['asian_est', 'white_est', 'black_est', 'api_est', 'hisp_est', 'nhopi_est'] #list of columns used in the function

    for i in qgwjsbgkjbkasbg:
        totalsum += df.at[505, i] #row number
        
    return df.at[row, column2]/totalsum # column divided by the sum of all of the columns 



asianperc = percentage(505, 'asian_est') #plug in row and column 
whiteperc = percentage(505, 'white_est')
blackperc = percentage(505, 'black_est')
apiperc = percentage(505, 'api_est')
hispperc = percentage(505, 'hisp_est')
nhopiperc = percentage(505, 'nhopi_est')

labels = ['Asian est', 'White est', 'Black est', 'Asian Pacific Islander est', 'Hispanic est', 'Native Hawaiian Other Pacific Islander est '] #label pi chart 
sizes = [asianperc, whiteperc, blackperc, apiperc, hispperc, nhopiperc] 

matplotlib.style.use('ggplot') #set stylesheet (find options here: https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)
matplotlib.rcParams['font.size'] = 43
matplotlib.rcParams['figure.figsize'] = (35, 35) #standardise figure size
matplotlib.rcParams['font.family'] =  'Verdana'

#plt.figure(figsize = (40,40))
plt.pie(sizes, autopct = '%1.1f%%', labels = labels) #plot labels 

```
