```python
import pandas as pd
import matplotlib.pyplot as plt
file = '2006-2011_race_and_ethncity_data_testing.csv' #import file
df = pd.read_csv(file)

df = df[df['Grade'] == '_All'] #limit df to all grades

dfAsian = df[df['Category'] == 'Asian'] 
df.head() 

def meanfinder(Race):
    dfRace = df[df['Category'] == Race] #setting what coulmn for the function
    return sum(dfRace['Mean_Scale_Score'])/len(dfRace['Mean_Scale_Score']) #find mean of column by adding them all together and dividing by the length

print(meanfinder('Asian')) #plug in race 
print(meanfinder('Black'))
print(meanfinder('Hispanic'))
print(meanfinder('White'))
```

```python
import pandas as pd #import libraries
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager
```

```python
#SET STYLE FOR REST OF NOTEBOOK
matplotlib.style.use('ggplot') #set stylesheet (find options here: https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)
matplotlib.rcParams['font.size'] = 23
matplotlib.rcParams['figure.figsize'] = (8.5, 8.5) #standardise figure size
matplotlib.rcParams['font.family'] =  'Verdana'
#matplotlib.rcParams['font.weight'] =  'bold' #<-consider adding this line (uncommented) if using a thin font
```

```python
#GET Fonts and write to file fonts.txt in same folder (NOTE: not every font will work)
fonts = font_manager.findSystemFonts(fontpaths = None, fontext = 'ttf')
file = open('fonts.txt', 'w')
file.write(str(fonts))
```

```python
def testraceyear(Race): #define function and choose column which is race
    meanyearlist = []

    for i in range(2006, 2012): #set range for the years in that dataframe that i'm using 
        meanyearlist.append(yearmeanfinder(i, Race)) 
        
    return meanyearlist
    
#print(agfiuawbguibuw('Asian'))

plt.plot(range(2006, 2012), testraceyear('Asian'), 'orange') #plug in the year and the race and choose color 
plt.plot(range(2006, 2012), testraceyear('Black'), 'cyan')
plt.plot(range(2006, 2012), testraceyear('White'), 'navy')
plt.plot(range(2006, 2012), testraceyear('Hispanic'), 'red')
plt.title('Mean Test Scores by Race from 2006-2011') #title scatter plot
plt.xlabel('Year') #label x-axis
plt.ylabel('Mean Test Score'); #label y-axis
```
